<h2 style="color:DodgerBlue">Описание проекта:</h2>

----

Создать базовый класс Book в C#, который будет представлять информацию о
книгах. На основе этого класса разработать 2-3 производных класса,
демонстрирующих принципы наследования и полиморфизма. В каждом из классов
должны быть реализованы новые атрибуты и методы, а также переопределены
некоторые методы базового класса для демонстрации полиморфизма.
Требования к базовому классу Book:
• Атрибуты: Название (Title), Автор (Author), Год издания (YearOfPublication).
• Методы:
o GetInfo(): метод для получения информации о книге в виде строки.
o Read(): метод для вывода сообщения о чтении книги.
o Borrow(): метод для вывода сообщения о выдаче книги на чтение.
Требования к производным классам:
1. Учебник (Textbook): Должен содержать дополнительные атрибуты, такие как
Предмет (Subject). Метод Read() должен быть переопределен для
добавления информации о предмете при чтении учебника.
2. Художественная литература (Fiction): Должен содержать дополнительные
атрибуты, такие как Жанр (Genre). Метод Borrow() должен быть
переопределен для добавления информации о жанре при выдаче книги на
чтение.
3. Научная литература (ScientificLiterature) (если требуется третий класс):
Должен содержать дополнительные атрибуты, такие как Область науки
(FieldOfScience). Метод GetInfo() должен быть переопределен для включения
информации об области науки в описании книги.

```
public interface IDigitalCopy { string FileFormat { get; } double FileSizeMb { get; } void Download(); }
public interface IAudioBook { int DurationMinutes { get; } void PlaySample(); }

public class Book
{
    public string Title { get; set; }
    public string Author { get; set; }
    public int YearOfPublication { get; set; }
    public bool IsAvailable { get; private set; } = true;
    public double Rating { get; private set; }

    public Book(string title="N/A", string author="N/A", int year=0) { Title=title; Author=author; YearOfPublication=year; }
    public virtual string GetInfo() => $"{Title} — {Author}, {YearOfPublication}";
    public virtual void Read() => Console.WriteLine($"Читаю \"{Title}\"");
    public void Read(string chapter) => Console.WriteLine($"\"{Title}\": {chapter}");
    public virtual void Borrow() { if(!IsAvailable){Console.WriteLine($"\"{Title}\" занята");return;} IsAvailable=false; Console.WriteLine($"Выдана: \"{Title}\""); }
    public void Borrow(int days){ Borrow(); Console.WriteLine($"Срок {days} дн."); }
    public void ReturnBook(){ IsAvailable=true; Console.WriteLine($"Возврат: \"{Title}\""); }
    public void Rate(double stars){ if(stars<0||stars>5) return; Rating = Rating==0? stars : Math.Round((Rating+stars)/2,2); Console.WriteLine($"Рейтинг \"{Title}\": {Rating}"); }
    public void Exchange(Book other)=>Console.WriteLine($"Обмен: \"{Title}\" <-> \"{other.Title}\"");
}

public class AcademicBook : Book
{
    public string Publisher { get; set; }
    public int Edition { get; set; }
    public AcademicBook(string t,string a,int y,string pub="Pub",int ed=1):base(t,a,y){ Publisher=pub; Edition=ed; }
    public override string GetInfo()=> base.GetInfo()+$" — {Publisher}, изд. {Edition}";
}

public class Textbook : Book, IDigitalCopy
{
    public string Subject { get; set; }
    public string FileFormat { get; } = "PDF";
    public double FileSizeMb { get; } = 8.0;
    public Textbook(string t,string a,int y,string subj):base(t,a,y){ Subject=subj; }
    public override void Read()=>Console.WriteLine($"Учебник \"{Title}\" по {Subject}");
    public void Download()=>Console.WriteLine($"Скачиваю \"{Title}\" ({FileFormat}, {FileSizeMb} МБ)");
}

public class Fiction : Book, IAudioBook
{
    public string Genre { get; set; }
    public int DurationMinutes { get; } = 600;
    public Fiction(string t,string a,int y,string g):base(t,a,y){ Genre=g; }
    public override void Borrow(){ Console.WriteLine($"Выдача художественной: \"{Title}\" ({Genre})"); base.Borrow(); }
    public void PlaySample()=>Console.WriteLine($"Фрагмент аудио: \"{Title}\"");
}

public class ScientificLiterature : AcademicBook, IDigitalCopy
{
    public string FieldOfScience { get; set; }
    public string FileFormat { get; } = "PDF";
    public double FileSizeMb { get; } = 5.5;
    public ScientificLiterature(string t,string a,int y,string field,string pub="Sci",int ed=1):base(t,a,y,pub,ed){ FieldOfScience=field; }
    public override string GetInfo()=> $"{Title} — {Author}, {YearOfPublication}. Область: {FieldOfScience}";
    public void Download()=>Console.WriteLine($"Скачиваю \"{Title}\" ({FileFormat}, {FileSizeMb} МБ)");
}

public class Library<T> where T: Book
{
    private readonly List<T> items = new();
    public void Add(T item)=>items.Add(item);
    public void Display()=>items.ForEach(x=>Console.WriteLine(x.GetInfo()));
    public void BorrowAll()=>items.ForEach(x=>x.Borrow());
}

var t = new Textbook("Математика 10","Иванов",2020,"Математика");
var f = new Fiction("Война и мир","Л. Толстой",1869,"Роман");
var s = new ScientificLiterature("Краткая история времени","С. Хокинг",1988,"Физика");

Console.WriteLine(t.GetInfo()); t.Read(); t.Borrow(30); t.Download();
Console.WriteLine(f.GetInfo()); f.Read(); f.Borrow(); f.PlaySample();
Console.WriteLine(s.GetInfo()); s.Read(); s.Download();

t.Exchange(f); t.Rate(5); t.Rate(3.5); t.ReturnBook();

var lib = new Library<Book>();
lib.Add(t); lib.Add(f); lib.Add(s);
lib.Display();
lib.BorrowAll();
```
