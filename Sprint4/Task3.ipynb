```
// ------------------------- ИНТЕРФЕЙС -------------------------
public interface IBook
{
    string GetInfo();
    void Read();
    void Borrow();
}

// ------------------------- БАЗОВЫЙ КЛАСС -------------------------
public abstract class Book : IBook
{
    // Основные атрибуты
    public string Title { get; set; }
    public string Author { get; set; }
    public int YearOfPublication { get; set; }

    // Дополнительные атрибуты
    public string Publisher { get; set; }
    public int Pages { get; set; }
    public double Rating { get; private set; }
    public bool IsAvailable { get; set; } = true;

    // Конструктор
    protected Book(string title, string author, int year, string publisher, int pages)
    {
        Title = title;
        Author = author;
        YearOfPublication = year;
        Publisher = publisher;
        Pages = pages;
    }

    // Методы базового класса
    public virtual string GetInfo()
    {
        return $"{Title} — {Author}, {YearOfPublication} г., издательство: {Publisher}, {Pages} стр.";
    }

    public virtual void Read()
    {
        Console.WriteLine($"Вы читаете книгу '{Title}' автора {Author}.");
    }

    public virtual void Borrow()
    {
        if (IsAvailable)
        {
            IsAvailable = false;
            Console.WriteLine($"Книга '{Title}' выдана на чтение.");
        }
        else
        {
            Console.WriteLine($"Книга '{Title}' уже находится у читателя.");
        }
    }

    // Полиморфизм: перегрузка метода Rate
    public void Rate(int rating)
    {
        Rating = rating;
        Console.WriteLine($"Книга '{Title}' получила оценку {rating}/5.");
    }

    public void Rate(double rating, string comment)
    {
        Rating = rating;
        Console.WriteLine($"Книга '{Title}' получила оценку {rating}/5. Комментарий: {comment}");
    }

    // Взаимодействие между объектами (например, сравнение книг)
    public bool CompareYear(Book other)
    {
        return this.YearOfPublication > other.YearOfPublication;
    }

    public void Return()
    {
        IsAvailable = true;
        Console.WriteLine($"Книга '{Title}' возвращена в библиотеку.");
    }
}

// ------------------------- ПРОИЗВОДНЫЙ КЛАСС 1: УЧЕБНИК -------------------------
public class Textbook : Book
{
    public string Subject { get; set; }
    public string EducationLevel { get; set; }
    public int Edition { get; set; }

    public Textbook(string title, string author, int year, string publisher, int pages, string subject, string educationLevel, int edition)
        : base(title, author, year, publisher, pages)
    {
        Subject = subject;
        EducationLevel = educationLevel;
        Edition = edition;
    }

    // Полиморфизм (перекрытие)
    public override void Read()
    {
        Console.WriteLine($"Чтение учебника '{Title}' по предмету '{Subject}' ({EducationLevel} уровень, {Edition}-е издание).");
    }

    // Перегрузка
    public void Read(bool aloud)
    {
        if (aloud)
            Console.WriteLine($"Учебник '{Title}' читается вслух для группы студентов.");
        else
            Read();
    }

    public void OpenChapter(int chapter)
    {
        Console.WriteLine($"Открыта глава {chapter} учебника '{Title}'.");
    }
}

// ------------------------- ПРОИЗВОДНЫЙ КЛАСС 2: ХУДОЖЕСТВЕННАЯ ЛИТЕРАТУРА -------------------------
public class Fiction : Book
{
    public string Genre { get; set; }
    public string Language { get; set; }
    public bool IsBestseller { get; set; }

    public Fiction(string title, string author, int year, string publisher, int pages, string genre, string language, bool isBestseller)
        : base(title, author, year, publisher, pages)
    {
        Genre = genre;
        Language = language;
        IsBestseller = isBestseller;
    }

    // Полиморфизм (перекрытие)
    public override void Borrow()
    {
        if (IsAvailable)
        {
            IsAvailable = false;
            Console.WriteLine($"Художественная книга '{Title}' жанра '{Genre}' выдана на чтение ({Language} язык).");
        }
        else
        {
            Console.WriteLine($"'{Title}' уже выдана.");
        }
    }

    // Перегрузка метода Borrow
    public void Borrow(string readerName)
    {
        if (IsAvailable)
        {
            IsAvailable = false;
            Console.WriteLine($"Читателю {readerName} выдана книга '{Title}' жанра '{Genre}'.");
        }
        else
        {
            Console.WriteLine($"'{Title}' уже у другого читателя.");
        }
    }

    public void Translate(string newLanguage)
    {
        Language = newLanguage;
        Console.WriteLine($"Книга '{Title}' переведена на язык: {Language}.");
    }
}

// ------------------------- ПРОИЗВОДНЫЙ КЛАСС 3: НАУЧНАЯ ЛИТЕРАТУРА -------------------------
public class ScientificLiterature : Book
{
    public string FieldOfScience { get; set; }
    public string DOI { get; set; }
    public int Citations { get; private set; }

    public ScientificLiterature(string title, string author, int year, string publisher, int pages, string field, string doi)
        : base(title, author, year, publisher, pages)
    {
        FieldOfScience = field;
        DOI = doi;
        Citations = 0;
    }

    // Полиморфизм (перекрытие)
    public override string GetInfo()
    {
        return $"{Title} — {Author}, {YearOfPublication} г., {FieldOfScience}. DOI: {DOI}, цитирований: {Citations}.";
    }

    public void Cite()
    {
        Citations++;
        Console.WriteLine($"Работа '{Title}' процитирована. Общее количество цитирований: {Citations}.");
    }

    // Перегрузка метода Cite()
    public void Cite(int times)
    {
        Citations += times;
        Console.WriteLine($"Работа '{Title}' получила {times} новых цитирований. Всего: {Citations}.");
    }
}

// ------------------------- GENERIC КЛАСС -------------------------
public class Library<T> where T : Book
{
    private List<T> books = new List<T>();

    public void AddBook(T book)
    {
        books.Add(book);
        Console.WriteLine($"Добавлена книга: {book.Title}");
    }

    public void ShowAllBooks()
    {
        Console.WriteLine("\nСписок книг в библиотеке:");
        foreach (var book in books)
            Console.WriteLine(book.GetInfo());
    }

    public void FindByAuthor(string author)
    {
        Console.WriteLine($"\nКниги автора {author}:");
        foreach (var book in books)
            if (book.Author == author)
                Console.WriteLine(book.Title);
    }
}

// ------------------------- УПРАВЛЕНИЕ ЗАВИСИМОСТЯМИ -------------------------
public class LibraryService
{
    private readonly IBook _book;

    public LibraryService(IBook book)
    {
        _book = book;
    }

    public void ProcessBook()
    {
        Console.WriteLine(_book.GetInfo());
        _book.Read();
        _book.Borrow();
    }
}

var textbook = new Textbook("Математика", "Иванов", 2021, "Просвещение", 350, "Алгебра", "Средняя школа", 3);
var fiction = new Fiction("Война и мир", "Л. Толстой", 1869, "Москва", 1225, "Роман", "Русский", true);
var science = new ScientificLiterature("Теория относительности", "А. Эйнштейн", 1916, "Springer", 240, "Физика", "10.1234/relativity.1916");

// Работа с generic библиотекой
var library = new Library<Book>();
library.AddBook(textbook);
library.AddBook(fiction);
library.AddBook(science);
library.ShowAllBooks();
library.FindByAuthor("Л. Толстой");

// Управление зависимостями
Console.WriteLine("\n--- Тест LibraryService ---");
var service = new LibraryService(textbook);
service.ProcessBook();

Console.WriteLine("\n--- Демонстрация полиморфизма ---");
textbook.Read();
textbook.Read(true);
fiction.Borrow("Алексей");
science.Cite(3);
```
